# Workflows with AWS Lambda

## 1. Lambda
1. Vào **Services** -> **Compute** -> **Lambda** -> **Create function**. Chọn **Runtime** là `Python`.

    <img src="images/aws1.png" width=1000>

2. Thay đổi code thành:
    ```python
    import json

    def lambda_handler(event, context):
        try:
            assert True==False
        except:
            raise
        return {
            'statusCode': 200,
            'body': json.dumps('Hello from Lambda!')
        }
    ```

    Ấn **Test** -> **Create new test event** -> Để default -> **Save** -> **Test** \
    **Chú ý**: Nếu thay đổi code và muốn test lại thì phải ấn **Deploy** trước.

    <img src="images/aws2.png" width=500>

    Sẽ thấy có error. Vào **Monitor** -> **View CloudWatch logs** để xem kĩ log.

    <img src="images/aws3.png" width=1500>

3. Vào **Configuration** -> **General configuration** -> **Edit**. Xem tên role rồi vào **IAM** thêm permission `S3FullAccess` cho role này.

    <img src="images/aws4.png" width=600>

4. Tạo Lambda function bằng code: Xem file `1_demo1_Lambda.ipynb`.

## 2. Triggering a Lambda Function
1. Vào **EventBrideg** -> **Buses** -> **Rules** -> **Create rule**. Chọn Rule type là **Schedule** -> **Continue**
    <img src="images/aws5.png" width=800>

2. Chọn **Occurence** là **Recurring schedule**, **Schedule type** là **Rate-based schedule**, **rate** là `1 minute`, **Flexible time window** là `Off` -> **Next**

    <img src="images/aws6.png" width=600>

3. Chọn **Target** là **AWS Lambda Invoke** -> Chọn function đã tạo -> **Next** -> **Next** -> **Create schedule**

    <img src="images/aws7.png" width=600>

   Giờ cứ mỗi 1 minute thì `lambda_function` sẽ đc triggered 1 lần. Xem kĩ ở logs.

4. Trigger Lambda function bằng code: Xem file `3_demo2_Triggering_Lambda.ipynb`. \
   Vào **Lambda** -> **Monitor** để kiểm tra.

## 3. Creating Workflows
1. Tạo thêm một Lambda function mà luôn `True`
    ```python
    import json

    def lambda_handler(event, context):
        try:
            assert True==True
        except:
            raise
        return {
            'statusCode': 200,
            'body': json.dumps('Hello from Lambda!')
        }
    ```

2. Vào **Step Functions** -> **State machines** -> **Create state machine** \
   Trong **Actions**, kéo thả 2 `AWS Lambda Invoke` vào, chọn hai Lambda functions luôn `False` và luôn `True` cho từng cái -> **Create**.

   <img src="images/aws8.png" width=1500>

3. Trong state machine vừa tạo, ấn **Start execution**. Theo dõi KQ.

   <img src="images/aws9.png" width=1500>

4. Tạo một state machine khác. Thiết kế như dưới.

   <img src="images/aws10.png" width=1500>

5. Vào **Error handing** -> **Add new catcher**. Trong **Errors**, chọn `States.ALL`. \
   Trong **Fallback state**, chọn **Add new state**. \
   Trong tab **Flow**, kéo thả `Success` vào đó. Thêm `Fail` vào nhánh còn lại. \
   Flow này nghĩa là nếu Lambda function fails thì ra state Success, còn nếu Lambda function success thì ra state Fail. \
   Ấn **Create**.

   <img src="images/aws11.png" width=1500>

6. Start execution và quan sát KQ.

   <img src="images/aws12.png" width=1500>

7. Để tạo State machine bằng code, trong state machine đã tạo, vào **Definition** để thấy code.

    <img src="images/aws13.png" width=1500>

8. Thay code đó vào phần `definition` trong file `5_demo3_Step_Functions.ipynb`.
